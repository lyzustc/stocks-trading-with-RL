```python
from dqn_algo import *
```

```python
dqn = DQNAlgo(filepath='./data/15.csv', netname='conv1d', algoname='duel')
```

```python
rewards, total_rewards = dqn.train(max_iteration=1e5, save_pth='./model.pth')
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(15,5))
fig.add_subplot(1,2,1)
plt.plot(list(range(len(rewards))), rewards)
plt.xlabel('iteration')
plt.ylabel('reward')
fig.add_subplot(1,2,2)
plt.plot(list(range(len(rewards))), total_rewards[1:])
plt.xlabel('iteration')
plt.ylabel('total reward')
plt.show()
```
