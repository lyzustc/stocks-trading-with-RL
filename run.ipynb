```python
from dqn_algo import *
```

```python
dqn = DQNAlgo(filepath='./data/15.csv', netname='fc', algoname='duel')
```

```python
rewards11, total_rewards11 = dqn.train(max_iteration=1e6, save_pth='./modelduelfc.pth')
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(15,5))
fig.add_subplot(1,2,1)
plt.plot(list(range(len(rewards11))), rewards11)
plt.xlabel('iteration')
plt.ylabel('reward')
fig.add_subplot(1,2,2)
plt.plot(list(range(len(rewards11))), total_rewards11[1:])
plt.xlabel('iteration')
plt.ylabel('total reward')
plt.show()
```
